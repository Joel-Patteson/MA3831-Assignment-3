```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
```

```python
url = "https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc"
```

```python
def all_page_link(start_url):
    all_urls = []
    url = start_url
    while(url != None):            #Loop around all the required webpages and terminates when last page arive!
        all_urls.append(url)
        soup = BeautifulSoup(requests.get(url).text,"html.parser")
        next_links = soup.find_all(class_='lister-page-next next-page')    #Extracts the next page link.
        if (len(next_links) == 0):         # If their is no next page, it returns 0.
            url = None
        else:
            next_page = "https://www.imdb.com" + next_links[0].get('href')
            url = next_page
    return all_urls
```

```python
main_array = []
for url in tqdm(all_page_link("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc")):     #Runs the function for all the pages.
    soup = BeautifulSoup(requests.get(url).text,"html.parser")         #Extracts out the main html code.
    for link in soup.find_all(class_='lister-item-content'):
        name = link.find('a').text
        year = years = link.h3.find('span', class_='lister-item-year').text
        run_time = link.find('span',{"class":"runtime"}).text
        genre = link.find('span',{"class":"genre"}).text[1:]
        director=link.find('p',class_='').find_all('a')[0].text
        about = link.find_all('p')[1].text[5:]
        rating = imdb = float(link.strong.text)
        list_of_all = [name,year,run_time,genre,director,about,rating]
        main_array.append(list_of_all)
```

```python
index = ["movie","year","run_time","genre","director","about","rating"]
```

```python
df = pd.DataFrame(main_array,columns=index) 
```
